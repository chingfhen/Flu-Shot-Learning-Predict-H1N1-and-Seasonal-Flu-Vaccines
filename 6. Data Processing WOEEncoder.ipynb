# About:

- this notebook applies WOEEncoder to train and test set

```python
import pandas as pd
train_features = pd.read_csv(r"C:\Users\tanch\Documents\Coding Competitions\DataDriven\Flu Shot Learning\local\data\training_set_features.csv",index_col = "respondent_id")
train_labels = pd.read_csv(r"C:\Users\tanch\Documents\Coding Competitions\DataDriven\Flu Shot Learning\local\data\training_set_labels.csv",index_col = "respondent_id")
test_features = pd.read_csv(r"C:\Users\tanch\Documents\Coding Competitions\DataDriven\Flu Shot Learning\local\data\test_set_features.csv",index_col = "respondent_id")
```

## Apply WOEEncoder - h1n1_vaccine
- encode all features including numerical ones with WOEEncoder
    - note: numerical features themselves had small number of numerical features, so they can be considered catgeorical

```python
y1 = train_labels['h1n1_vaccine']
```

```python
# fit on train set
from category_encoders.woe import WOEEncoder
woe = WOEEncoder(cols = train_features.columns)
woe.fit(train_features,y1)
```

```python
# transform train and test set
train_features_encoded_h1n1_vaccine = woe.transform(train_features)
test_features_encoded_h1n1_vaccine = woe.transform(test_features)
```

```python
train_features_encoded_h1n1_vaccine
```
